```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = 'Data/Analisis de medios_CPAC_2025.csv'
df = pd.read_csv(file_path)

# Display all rows
print(f"Total rows: {len(df)}")
display(df)
```

```python
# Count the number of articles per media outlet
counts = df['1. Nombre del medio'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel('Nombre del medio')
plt.ylabel('Numero de articulos')
plt.title('Distribucion de articulos por medio')

# Add data labels on top of each bar
for i, count in enumerate(counts.values):
    plt.text(i, count + 0.1, str(count), ha='center')
plt.tight_layout()
plt.show()
```

```python
# Count the number of articles per media outlet
counts = df['8. Valoración general de la nota'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel(' ')
plt.ylabel('Numero de articulos')
plt.title('Valoración general de la nota')

# Add data labels on top of each bar
for i, count in enumerate(counts.values):
    plt.text(i, count + 0.1, str(count), ha='center')
plt.tight_layout()
plt.show()
```

```python
# Count the frequency of each value in '9.1 ¿Cuáles?' column
counts = df['9.1 ¿Cuáles?'].value_counts()

# Create a bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.title('Distribución de respuestas para "9.1 ¿Cuáles?"')

# Add data labels on top of each bar
for i, count in enumerate(counts.values):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.tight_layout()
plt.show()
```

```python
# Define all possible options
options = [
    'Banderas', 
    'Fotos/ imágenes / nombres de actores políticos', 
    'Colores', 
    'Logo de partidos', 
    'Logos de colectivos', 
    'Fichas de desaparecidos', 
    'Otros ¿cuáles?'
]

# Initialize a dictionary to count occurrences of each option
option_counts = {option: 0 for option in options}

# Count occurrences of each option
for response in df['9.1 ¿Cuáles?'].dropna():
    # Split the response by semicolons and strip whitespace
    choices = [choice.strip() for choice in str(response).split(';')]
    
    # Increment the count for each option found in this response
    for option in options:
        if any(option in choice for choice in choices):
            option_counts[option] += 1

# Convert to DataFrame for easier plotting
results_df = pd.DataFrame({
    'Option': list(option_counts.keys()),
    'Count': list(option_counts.values())
})

# Sort by count in descending order
results_df = results_df.sort_values('Count', ascending=False)

# Create the bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='Option', y='Count', hue='Option', data=results_df, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel('Opción')
plt.ylabel('Frecuencia')
plt.title('Distribución de opciones en la pregunta "9.1 ¿Cuáles?"')

# Add data labels on top of each bar
for i, count in enumerate(results_df['Count']):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.tight_layout()
plt.show()
```

```python
# Define all possible options for collectives
options = [
    'Frente Feminista de Jalisco',
    'YoVoy8M',
    'Tapatías discas',
    'Familias de personas desaparecidas y víctimas de feminicidio',
    'Contingente carriola',
    'Mixto',
    'Bloque negro',
    'Batukada',
    'Contingente general de mujeres y disidencias'
]

# Initialize a dictionary to count occurrences of each option
option_counts = {option: 0 for option in options}

# Count occurrences of each option
for response in df['10.1 ¿Cuál colectivo se menciona en la nota? '].dropna():
    # Split the response by semicolons and strip whitespace
    choices = [choice.strip() for choice in str(response).split(';')]
    
    # Increment the count for each option found in this response
    for option in options:
        if any(option in choice for choice in choices):
            option_counts[option] += 1

# Convert to DataFrame for easier plotting
results_df = pd.DataFrame({
    'Option': list(option_counts.keys()),
    'Count': list(option_counts.values())
})

# Sort by count in descending order
results_df = results_df.sort_values('Count', ascending=False)

# Create the bar chart
plt.figure(figsize=(16, 8))
sns.barplot(x='Option', y='Count', hue='Option', data=results_df, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel('Colectivo')
plt.ylabel('Frecuencia')
plt.title('Distribución de colectivos mencionados en la nota')

# Add data labels on top of each bar
for i, count in enumerate(results_df['Count']):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.tight_layout()
plt.show()
```

```python
# Count the frequency of each value in the actor column
counts = df['13.1 ¿Quién es el actor involucrado en el que se enfoca la pieza? '].value_counts()

# Create a bar chart
plt.figure(figsize=(14,5))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, legend=False, palette="viridis")
plt.xticks(rotation=45, ha='right')

# Set y-axis ticks to increment by 1
max_count = counts.max()
plt.yticks(range(0, max_count + 1, 1))

# Add labels and title
plt.xlabel('Actor involucrado')
plt.ylabel('Frecuencia')
plt.title('¿Quién es el actor involucrado en el que se enfoca la pieza?')

plt.tight_layout()
plt.show()
```
